```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functions
plt.plot(x,y)
plt.xlabel('x Label')
plt.ylabel('y lable')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')

#plt.subplot(1,2,2)
#plt.plot(x,x,'b')
```

```python
#OO

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('x Label')
axes.set_ylabel('y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('x Label')
axes.set_ylabel('y Label')
axes.set_title('Title')

axes = fig.add_axes([0.2,0.5,0.4, 0.3])
axes.plot(y,x)
axes.set_xlabel('x iLabel')
axes.set_ylabel('y iLabel')
axes.set_title('iTitle')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#for current_ax in axes:
#current_ax.plot(x,y)
#axes.plot(x,y)
#using index
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2), dpi=100)  #dot per inch or picture per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1,figsize=(7,2), dpi=100)  #dot per inch or picture per inch
#ax = fig.add_axes([0,0,1,1])
ax[0].plot(x,y)
ax[1].plot(x,y)
fig.savefig('matplotgraph.png', dpi=200)
```

```python
#how t o ad legend
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='x Square')
ax.plot(x,x**3, label='y cubed')
ax.legend()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color="purple", lw=3, alpha=1, ls='dotted', marker='o', markersize=10,
       markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='green'
       )   # Color can be added in #rbg format, line width can be set in plot


#ax.plot(x,y,color='purple',lw=2,ls='dashed')

ax.set_xlim([0,1])  #Set x limit for plot
ax.set_ylim([0,2])  #Set Y limit for plot
```
